```python
# 데이터를 훈련과 테스트로 분할
# Random_state : 무작위에 같은 seed를 주어 실험때마다 다른 값이 되는 것을 방지

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
# X : 데이터[행렬] -> 대문자
# y : 레이블/타겟[벡터] -> 소문자

print("X_train 크기 : {}".format(X_train.shape))
print("y_train 크기 : {}".format(y_train.shape))
```

```python
# Train : Test = 75 : 25

print("X_test 크기 : {}".format(X_test.shape))
print("y_test 크기 : {}".format(y_test.shape))
```

```python
# tset_size만 변경
# Train : Test = 67 : 33

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],test_size=0.33,random_state=0)
print("X_train 크기 : {}".format(X_train.shape))
print("y_train 크기 : {}".format(y_train.shape))

print("X_test 크기 : {}".format(X_test.shape))
print("y_test 크기 : {}".format(y_test.shape))
```

```python
# random_state만 변경
# Train : Test = 75 : 25

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=21)
print("X_train 크기 : {}".format(X_train.shape))
print("y_train 크기 : {}".format(y_train.shape))

print("X_test 크기 : {}".format(X_test.shape))
print("y_test 크기 : {}".format(y_test.shape))
```

```python
# tset_size와 random_state 변경
# Train : Test = 60 : 40

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],test_size=0.40,random_state=42)
print("X_train 크기 : {}".format(X_train.shape))
print("y_train 크기 : {}".format(y_train.shape))

print("X_test 크기 : {}".format(X_test.shape))
print("y_test 크기 : {}".format(y_test.shape))
```

```python
# X_train 데이터를 사용해서 데이터 프레임 생성
# 열의 이름은 iris_dataset.feature_names에 있는 문자열 사용

# Dataframe : 2차원 테이블 데이터 구조를 가지는 자료형
# DataFrame = pandas.DataFrame(data, index=index, columns=columns)

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# 데이터프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬 생성

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o',
                           hist_kwds={'bins':20}, s=60, alpha=8, cmap=mglearn.cm3)
```

```python
# KNN분류기 불러오기
from sklearn.neighbors import KNeighborsClassifier

# 이웃의 수 k = 1
knn = KNeighborsClassifier(n_neighbors=1)

# 훈련데이터와 레이블로 모델 훈련
knn.fit(X_train, y_train)
```

```python
# 새로운 데이터로 분류해보기

X_new = np.array([[5,2.9,1,0.2]])
print("X_new.shape :", X_new.shape)
```

```python
# predict() : 해당 데이터로 예측

prediction = knn.predict(X_new)
print("예측 : ",prediction)
print("예측한 타깃의 이름 :",iris_dataset['target_names'][prediction])
```

```python
y_pred=knn.predict(X_test)
print("테스트 세트에 대한 예측값  : \n",y_pred)
```

```python
print("테스트 세트의 정확도 : {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
# score() : 생성된 모델의 성능 평가

print("테스트 세트의 정확도 : {:.2f}".format(knn.score(X_test,y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)

print("테스트 세트의 정확도 : {:.2f}".format(knn.score(X_test,y_test)))
```

```python
# k = 2

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)

print("테스트 세트의 정확도 : {:.2f}".format(knn.score(X_test,y_test)))
```
